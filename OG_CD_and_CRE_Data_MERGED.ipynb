# This notebook contains the process of merging original covid data we acquired with the CRE data and any additional adjusting for aesthetic purposes


```python
# Reading in data

cd = pd.read_csv('Covid_Data_Original_Processing_Complete.csv')
cr = pd.read_csv('CRE_Equity_County_19_Processing_Complete.csv')
```

```python
# Merging dataframes based on "county"

cdOgcr =pd.concat([cd,cr], axis = 1)

```

```python
cdOgcr
```

```python
# Dropping the subsequent unnamed column

cdOgcr = cdOgcr.drop('Unnamed: 0', axis = 1)
    
    
```

```python
# Dropping county cols to reinsert bccccc whenconcat it merged lol

cdOgcr = cdOgcr.drop(cdOgcr.columns[[1]], axis = 1)

```

```python
# Reinserting county

cdOgcr[['County']] = cr[['County']]
```
